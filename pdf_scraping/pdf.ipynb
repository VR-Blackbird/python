```python
from bs4 import BeautifulSoup as soup
from multiprocessing import Pool
import io
import os
import requests
from urllib.request import urlopen
```

```python
pdf_url = "https://www.rgpvonline.com/rgpv-first-year.html"
```

```python
pdf_links = set()
```

```python
pd_l = requests.get(pdf_url).content
page_soup = soup(pd_l)
page_soup
```

```python
all_pages = page_soup.select("p > a")
```

```python
for page in all_pages:
    if "html" in page['href']:
        pdf = page['href'][:-4] + 'pdf'
        pdf_links.add(pdf)
```

```python
len(pdf_links)
```

```python
folder = "pdf_scrap"
def download_pdfs(pdf, idx):
    pdf_req = requests.get(pdf).content
    path = folder + '/' + f"pdf-{idx+1}.pdf"
    with open(path, 'wb') as f:
        f.write(pdf_req)


if not os.path.exists(folder):
    os.mkdir(folder)

with Pool(processes=30) as pool:
    pool.starmap(download_pdfs, [(pdf, idx) for idx, pdf in enumerate(pdf_links)])
```
