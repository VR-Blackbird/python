```python
ones = np.ones((97, 1))
x_values = np.concatenate((ones, x_values), axis=1)
```

# Hypothesis function

```python
def hypothesis(x_values, initial_theta):
    return np.matmul(x_values,  initial_theta)
```

# Cost function

```python
def cost(x_values, theta, m, y):
    m = 97
    hyp = hypothesis(x_values, theta)
    return (1/2 * m) * sum((hyp - y)**2)
    
```

# Gradient Descent

```python
def gradient(x_values, y, m, theta, alpha, iterations):
    cost_history = []
    for i in range(iterations):
        hyp = hypothesis(x_values, theta)
        theta -= (alpha/m) * (np.matmul(np.transpose(x_values), hyp-y))
        cost_history.append(cost(x_values, theta, m, y))
    return theta, cost_history
```

```python
alpha = 0.01
iterations = 2000
m = 97
```

# Prediction

```python
parameters, cost_history = gradient(x_values, y_values, m, initial_theta, alpha, iterations)
```
