```python
import numpy as np

def r_matrix(x, a, h, k):
    x_matrix = np.vstack([1/x, -np.ones_like(x), np.ones_like(x)])
    coef_matrix = np.array([a, h, k])
    result = np.dot(coef_matrix, x_matrix)
    return result

# Example usage:
x_values = np.array([1, 2, 3, 4, 5])
a_value = 2
h_value = 1
k_value = 3

result = r_matrix(x_values, a_value, h_value, k_value)
print(result)
```
